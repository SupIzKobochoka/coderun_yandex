```python
import numpy as np
import pandas as pd

import statsmodels.api as sm
```

```python
df = pd.read_csv('data.csv', names=['x', 'f'])
df
```

```python
# Раскрыл скобки на листочке и получил линейную модель без константы
# Пытался вывести ошибки в отдельные значения, но ничего не получилось, пришлось коэффициенты оценивать с ошибками

df['sinx ** 2'] = np.sin(df['x']) ** 2
df['lnx ** 2'] = np.log(df['x']) ** 2
df['sinx lnx'] = np.sin(df['x']) * np.log(df['x'])
df['x ** 2'] = df['x'] ** 2
df
```

```python
model = sm.OLS(endog=df['f'], exog=df[['sinx ** 2', 'sinx lnx', 'lnx ** 2', 'x ** 2']])
model = model.fit()
model.summary()
```

```python
model.params[['sinx ** 2', 'lnx ** 2']] ** (1 / 2)
```
